```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'data/groovy-1_5_7.csv'
data = pd.read_csv(file_path)
```

```python
# Display basic information
print("Shape of the dataset:", data.shape)
print("\nData Types:\n", data.dtypes)
print("\nMissing Values:\n", data.isnull().sum())

# Display the first few rows of the dataset
print("\nFirst few rows of the dataset:\n", data.head())

# Generate descriptive statistics
print("\nDescriptive Statistics:\n", data.describe())
data.RealBug.value_counts()
```

```python
# Visualize the distribution of numerical features
numerical_features = data.select_dtypes(include=['int64', 'float64']).columns
print(len(numerical_features))
data[numerical_features[:10]].hist(bins=15, figsize=(15, 10), layout=(5, 5))
plt.tight_layout()
plt.show()

# Visualize the correlation matrix
plt.figure(figsize=(15, 10))
correlation_matrix = data[numerical_features[:10]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
